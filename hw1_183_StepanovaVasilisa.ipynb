```python
import collections
import json
import re
from string import punctuation
```

# функция для удаления пунктуации

```python
def strip_punc(s): 
    return ''.join(c for c in s if c not in punctuation)
```

# начальные переменные

```python
lang_counter = collections.Counter()  # переменная для подсчета языков
user_data = collections.Counter()  # для подсчета твитов у пользователя
tweet_autor = collections.Counter()  # для подсчета количества подписчиков
tweet_source = collections.Counter()  # для подсчета источников
hashtags_counter = collections.Counter()  # для подсчетов хэштэгов
full_text = []  # список всех твитов
original_tweet = []  # список оригинальных твитов
user_data_id = 0  # количество пользователей, у которых больше одного твита
delete_twitter = 0  # подсчет количества удаленных твитов
dictionary = {}  # частотный словарь
```

# заполнение списка твитами из файла 

```python
for line in open('hw_3_twitter.json', 'r', encoding='utf-8'):
    full_text.append(json.loads(line))
```

# выполняем цикл для каждого твита, подсчитываем требуемые значения в соотвествии с заданием

```python
for jsdata in full_text:  
    for key in jsdata.keys():
        if key == 'delete':  # считаем кол-во удаленных твитов
            delete_twitter += 1
        if key == 'lang':  # подсчиитываем список использувемых языков
            lang_counter[jsdata[key]] += 1
        if key == 'user':  #  считаем твиты у пользователей
            user_data[jsdata[key]['id']] += 1
        if key == 'entities':  # собираем хэштэги
            hashtags = jsdata[key]['hashtags']
            for tag in hashtags:
                hashtags_counter[tag['text']] += 1
        if key == 'retweeted' and not jsdata['retweeted'] and jsdata['lang'] == 'en':  # формируем список с оригинальными твитами на английском языке
            text = ascii(strip_punc(jsdata['text']).lower())
            original_tweet.append(text)
        if key == 'user':  # подсчитываем кол-во подписчиков для автора
            u = jsdata['user']
            tweet_autor[ascii(u['name'])] += u['followers_count']
        if key == 'source':  # собираем имточники из гипер линка
            source = jsdata[key]
            match = re.search(r">(.+?)<", source)
            if match:
                tweet_source[match.group(1)] += 1
```

# создание частотного словаря

```python
for key in original_tweet:
    if key in dictionary:
        value = dictionary[key]
        dictionary[key] = value + 1
    else:
        dictionary[key] = 1
```

# цикл для подсчета пользователей, у который больше одного твита

```python
for login in user_data:
    if user_data[login] > 1:
        user_data_id += 1
```
